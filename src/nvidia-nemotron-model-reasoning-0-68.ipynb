```r
get_model_versions <- function(base_path) {
  list(
    v1 = file.path(base_path, "600-samples-packing-false/1"),
    v2 = file.path(base_path, "1800s-lora-rank32-false/1"),
    v3 = file.path(base_path, "2400-1e-4_lr-all_linear-packingfalse/1"),
    v4 = file.path(base_path, "9500s-batch1-lr1e-4/1"),
    v5 = file.path(base_path, "cot-labels-3000samples/1"),
    v6 = file.path(base_path, "1200samples-cot-5e-5/1"),
    v7 = file.path(base_path, "1200samples-cot-/1")
  )
}

validate_path <- function(path) {
  if (!dir.exists(path)) stop(sprintf("Invalid path: %s", path))
  normalizePath(path)
}

zip_model <- function(path, output = "submission.zip") {
  path <- validate_path(path)
  system2("zip", args = c("-r", output, path))
  output
}

select_model <- function(models, key) {
  if (!key %in% names(models)) stop("Invalid model key")
  models[[key]]
}

run <- function(base_path, version_key, output = "submission.zip") {
  models <- get_model_versions(base_path)
  model_path <- select_model(models, version_key)
  zip_model(model_path, output)
}

base_path <- "/kaggle/input/models/kienngx/nemotron-nano-30b-trained/transformers"
run(base_path, "v7")
```
